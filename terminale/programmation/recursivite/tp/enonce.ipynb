## Les Fractales

# 1. Contexte

L'objectif de cette activité est de s'intéresser aux __fractales__.

Une __fractale__ est un objet géométrique «infiniment morcelé» dont des détails sont observables à une échelle arbitrairement choisie.

![](https://upload.wikimedia.org/wikipedia/commons/thumb/5/5e/Romanesco_broccoli_%28Brassica_oleracea%29.jpg/300px-Romanesco_broccoli_%28Brassica_oleracea%29.jpg)

<center>Le chou romanesco, un exemple de forme fractale naturelle, Source : Wikipedia</center>

![](https://upload.wikimedia.org/wikipedia/commons/thumb/b/b5/Mandel_zoom_04_seehorse_tail.jpg/320px-Mandel_zoom_04_seehorse_tail.jpg)

<center>Exemple de figure fractale, Source : Wikipedia</center>

En zoomant sur une partie de la figure, on peut retrouver toute la figure, on dit qu’elle est auto similaire. Autrement-dit, les fractales présentent une structure récursive !!

Même si un certain nombre de choses était déjà connu, on attribue la découverte des fractales à un polytechnicien français, _Benoît Mandelbrot_ (1924 - 2010).

## 2. Mode opératoire

Nous allons utiliser la bibliothèque Python turtle, qui permet de réaliser des dessins en déplaçant une tortue.

Un exemple d'utilisation de la bibliothèque :

```python
import turtle

# positionnement de la tortue
turtle.goto(0, 0)
turtle.speed("normal")

# Réalisation d'un triangle
for _ in range(3):
    turtle.forward(100)
    turtle.right(120)

turtle.penup()
turtle.goto(0, 0)
turtle.left(210)
turtle.pendown()

for _ in range(3):
    turtle.circle(50, 180)
    turtle.right(60)

# arrêt de la tortue
turtle.done()
```
