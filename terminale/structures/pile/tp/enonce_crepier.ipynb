# TP - Crêpier psychorigide



## 1. Rappel

A la fin de sa journée, un crêpier dispose d’une pile de crêpes désordonnée.

Le crêpier étant un peu psychorigide, il décide de ranger sa pile de crêpes, de la plus grande (en bas) à la plus petite (en haut).

Pour cette tâche, le crêpier peut faire une seule action : __glisser__ sa spatule entre deux crêpes et __retourner__ le haut de la pile.

## 2. Objectifs

L'objectif de ce TP est d'implémenter l'algorithme de tri du crêpier psychorigide.

Nous allons créer un module `crepier`, qui sera enrichi au fur et à mesure de la séance.

L'interface sera simple puisque seulement 3 fonctions sont nécessaires : retourner un sous-ensemble d'une pile, rechercher le maximum d'une pile et le tri crêpier à proprement parlé.

- `retourner(p, i)` : retourne l'ensemble des éléments d'une pile p de son sommet à l'indice i,
- `recherche_maximum(p, i)` : recherche le maximum des éléments d'une pile p de son sommet à l'indice i,
- `tri_crepier(p)` : effectue le tri crêpier en place sur la pile p.
