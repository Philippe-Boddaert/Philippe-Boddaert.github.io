# TP - Universalité des fonctions NAND et NOR

## 1. Contexte

La fonction NON-ET (NAND) est dite « universelle » (comme la fonction NON-OU), car elle permet de __reconstituer toutes les autres fonctions__ logiques (NON, ET, OU...).

| Symbole NAND | Symbole NOR |

De plus, son circuit électronique étant des plus simples, la fonction NON-ET sert souvent de « brique de base » à des circuits intégrés beaucoup plus complexes.

## 3. La fonction universelle NAND

Ci-dessous, la fonction python qui implémente la fonction NAND.

```python
def NAND(a, b):
    '''
    Fonction Logique NAND (NON-ET)
    :param a: (int) une variable booléenne sous la forme d'un entier 0 = False, 1 = True
    :param b: (int) une variable booléenne sous la forme d'un entier 0 = False, 1 = True
    :return: (int) le résultat de a NAND b
    :Exemples:
        >>> NAND(0, 0)
        1
        >>> NAND(0, 1)
        1
        >>> NAND(1, 0)
        1
        >>> NAND(1, 1)
        0
    '''
    return 1 if a + b == 0 else (a + b) % 2
```

## 4. La fonction universelle NOR

Ci-dessous, la fonction python qui implémente la fonction NOR.

```python
def NOR(a, b):
    '''
    Fonction Logique NOR (NON-OU)
    :param a: (int) une variable booléenne sous la forme d'un entier 0 = False, 1 = True
    :param b: (int) une variable booléenne sous la forme d'un entier 0 = False, 1 = True
    :return: (int) le résultat de a NOR b
    :Exemples:
        >>> NOR(0, 0)
        1
        >>> NOR(0, 1)
        0
        >>> NOR(1, 0)
        0
        >>> NOR(1, 1)
        0
    '''
    return 1 if a + b == 0 else 0
```
