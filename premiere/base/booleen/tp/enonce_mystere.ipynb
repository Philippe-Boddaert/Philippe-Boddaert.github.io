# TP - Mystère et bool de gomme



## 1. Contexte

L'idée de ce TP est de retrouver des expressions booléennes à partir de leur table de vérité.

A chaque fonction `mystere` suivante, sont données des doctests.
Ces doctests correspondent aux valeurs des tables de vérités de la fonction.

Votre objectif est de programmer le corps des fonctions booléennes en partant des tables de vérité et en utilisant les opérateurs booléens (`and`, `or`, `not` et `^`).
