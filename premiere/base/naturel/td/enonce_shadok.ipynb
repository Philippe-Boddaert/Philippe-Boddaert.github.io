# TD - Les Shadoks

## 1. Contexte

Les Shadoks sont des créatures anthropomorphes _fictives_, à l'apparence d'oiseaux rondouillards, avec de longues pattes filiformes, des ailes minuscules et préhensiles, et de rares cheveux.



Les Shadoks réussissent à compter avec seulement 4 mots de leur bien maigre vocabulaire : __GA BU ZO__ et __MEU__.

Dans cette [vidéo](https://youtu.be/lP9PaDs2xgQ), le professeur Shadoko explique simplement la méthode employée par les Shadoks.

Ci dessous, l'explication du professeur Shadoko:

_"Notez bien. Quand il n'y a pas de Shadok, on dit __GA__ et on écrit __O__._

_Quand il y a un Shadok en plus on dit __BU__ et on écrit __I__._

_Quand il y a un Shadok en plus on dit __ZO__ et on écrit __L__._

_Quand il y a un Shadok en plus on dit __MEU__ et on écrit __◺__._

_Quand il y a un Shadok en plus on se rend compte qu'on ne peut plus compter._

_On met tous les Shadoks dans une poubelle et on a __BU__ poubelle et __GA__ Shadok._

_Et on continue: __BUGA__; __BUBU__; __BUZO__...Et on continue"_
