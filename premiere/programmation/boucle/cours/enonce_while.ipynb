## 1. Définition

Souvent, dans un programme, il est nécessaire de répéter un bloc d'instructions sans connaitre le nombre d'itérations. Pour cela, on utilise ce qu'on appelle des __boucles non bornées__.

## 2. Construction d'une boucle non bornée

En Python, les boucles non bornées (dites Tant que) se codent en utilisant l'instruction `while` :

```python
while condition:
    bloc_instructions
```

__Remarques__ :

- `condition` est une __variable ou expression booléenne__, qui est soit vraie (`True`) soit fausse (`False`),
- Tant que `condition` vaut `True` les instructions du bloc sont répétées,
- On passe à la suite du programme dès que condition vaut `False`,
- Ne pas oublier les deux points à la fin de la ligne avec while qui permettent d'ouvrir le bloc d'instructions à exécuter (à répéter),
- les instructions à effectuer sont indentées d'une tabulation par rapport au `while`.

## 3. Les pièges des boucles non bornées

Les boucles bornées (`FOR`) sont exécutées un nombre de fois fixe et pré-défini.

Les boucles non bornées (`WHILE`) sont exécutées un nombre de fois non défini, ce qui peut avoir des effets de bords indésirables.

Ainsi, 2 pièges sont à éviter.

### 3.1. piège n°1 : ne JAMAIS ENTRER dans la boucle

Soit Le programme suivant :

```python
a = 0

while a > 10:
    print("ce texte ne s'écrira jamais")
    a = a + 1

print("fini") 
```
